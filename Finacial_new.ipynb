#### (d) Come up with any one KPI which can help define the best stock across different Sector, you may need to learn a little bit of Finance for the same

```python
growth = df[ (df["1-Year Return(%)"] <df["3-Year Return"]) &  (df["3-Year Return"] <df["5-Year Return(%)"])  &
            (df["5-Year Return(%)"] <df["10-Year Return(%)"]) ]

one_to_ten = {"Company":growth["Company"], "Sector":growth["Sector"],
               "1 to 3year growth": (growth["3-Year Return"] - growth["1-Year Return(%)"]),
               "3 to 5year growth" : (growth["5-Year Return(%)"] - growth["3-Year Return"]),
               "5 to 10year growth" : (growth["10-Year Return(%)"] - growth["5-Year Return(%)"]) } 

one_to_ten["Total Growth"] = one_to_ten["1 to 3year growth"] + one_to_ten["3 to 5year growth"] + one_to_ten["5 to 10year growth"]

growth_df = pd.DataFrame(one_to_ten)


growth_df.sort_values(by = ["Total Growth"], ascending=False)


kpi = growth_df.groupby("Sector")["Company", "Sector", "Total Growth"].max()
kpi
```
